```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data =pd.read_csv("student.info.csv")
```

```python
plt.scatter(x=data.StudyHours,y=data.Marks,)
plt.xlabel("Study_hours")
plt.ylabel("Marks")
plt.title("scatter plot")
plt.show()
```

```python
data.isnull().sum()
```

```python
x=data.drop("Marks",axis="columns")
y=data.drop("StudyHours",axis="columns")
print("shape of  X  = ",x.shape)
print("shape of Y=",y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
from os import XATTR_CREATE
print("shape of  X_train = ",x_train.shape)
print("shape of y_train=",y_train.shape)
print("shape of  X_test  = ",x_test.shape)
print("shape of y_test=",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
m=9.78
c=0.98
y=m*4 +c
y
```

```python
lr.predict([[4]])[0][0].round(2)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
pd.DataFrame(np.c_[x_test,y_test,y_pred],columns=["study_hours","Student_marks_orig","student_pred"])
```

```python
lr.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_train,lr.predict(x_train),color='r')
```

```python
import joblib
joblib.dump(lr,"students_marks_predictor_model.pkl")
```
